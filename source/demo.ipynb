# Importing Data 

```python
import pandas as pd


data = pd.read_csv("HCMcity_temperature.csv")
data.head()
```

```python
data.info()
```

```python
data_model = data[data['LOC'] == 'ThuDuc']
```

```python
data_model = data_model[['WS50M', 'WS50M_RANGE', 'QV2M', 'PRECTOTCORR',
'T2M_RANGE', 'ALLSKY_SFC_SW_DWN', 'T2M', 'ALLSKY_KT',
'ALLSKY_SFC_LW_DWN', 'WS2M', 'ALLSKY_SFC_PAR_TOT', 'LOC']]
```

```python
data_model.head()
```

# Creating X and Y data frame for Training and Testing Purpose

```python
from sklearn.model_selection import train_test_split

X = data_model.drop(['T2M','LOC'], axis = 1)
X.head()
```

```python
Y = data_model['T2M']
Y.head()
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X.iloc[:, :5] = sc.fit_transform(X.iloc[:, :5])
```

# Splitting dataset for Training and Testing Purpose

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, random_state=0)
print(X.shape, Y.shape)
```

```python
data_model.info()
```

# Modeling

## Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
lreg = LinearRegression()
lreg.fit(X_train, y_train)
```

## Support Vector Regression

```python
from sklearn.svm import SVR
svm_rbf = SVR(C = 1000, gamma = 0.001, kernel = 'rbf') 
svm_rbf.fit(X_train, y_train)
```

## Random Forest Regression Model

```python
from sklearn.ensemble import RandomForestRegressor
rf= RandomForestRegressor(bootstrap=False, max_depth=None, max_features='sqrt', 
                            min_samples_leaf=1, min_samples_split=10, n_estimators=150, random_state=0)
rf.fit(X_train, y_train)
```

# Test model

```python
X_test.head()
```

```python
y_test.head()
```

```python
X_test_demo = X_test.iloc[5:39,:]
Y_test_demo = y_test.iloc[5:39]
```

```python
lpred = lreg.predict(X_test_demo)
svm_pred = svm_rbf.predict(X_test_demo)
rf_pred=rf.predict(X_test_demo)
```

```python
df_result = pd.DataFrame({'Y_test_TRUE': Y_test_demo,'LinearRegression':lpred, 'SVM':svm_pred, 'RandomForest':rf_pred})
df_result
```
