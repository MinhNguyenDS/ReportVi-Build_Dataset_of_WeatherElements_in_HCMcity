### KIỂM TRA VÀ XỬ LÝ CÁC DỮ LIỆU BỊ NHIỄU (uotlier)

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Kiem tra cac outlier theo moi cot bang do thi tu phan vi
fig,((ax1,ax2,ax3), (ax4,ax5,ax6), (ax7,ax8,ax9), (ax10,ax11,ax12)) = plt.subplots(figsize=(20,15),ncols=3,nrows=4)
textprops = {"fontsize":20}

sns.boxplot(x = data_HCM['WS50M'], ax = ax1)
sns.boxplot(x = data_HCM['WS50M_RANGE'], ax = ax2)
sns.boxplot(x = data_HCM['QV2M'], ax = ax3)
sns.boxplot(x = data_HCM['PRECTOTCORR'], ax = ax4)
sns.boxplot(x = data_HCM['T2M_RANGE'], ax = ax5)
sns.boxplot(x = data_HCM['ALLSKY_SFC_SW_DWN'], ax = ax6)
sns.boxplot(x = data_HCM['T2M'], ax = ax7)
sns.boxplot(x = data_HCM['ALLSKY_KT'], ax = ax8)
sns.boxplot(x = data_HCM['ALLSKY_SFC_LW_DWN'], ax = ax9)
sns.boxplot(x = data_HCM['WS2M'], ax = ax10)
sns.boxplot(x = data_HCM['ALLSKY_SFC_PAR_TOT'], ax = ax11);
```

=> Xóa các outlier nằm ngoài phân phối 

```python
import numpy as np
out_col = {'WS50M':1, 'WS50M_RANGE':2,
       'QV2M':3, 'PRECTOTCORR':4, 'T2M_RANGE':5, 'ALLSKY_SFC_SW_DWN':6,
       'T2M':7, 'ALLSKY_KT':8, 'ALLSKY_SFC_LW_DWN':9, 'WS2M':10, 'ALLSKY_SFC_PAR_TOT':11}

for variable in out_col.keys():
    q75, q25 = np.percentile(data_HCM[variable], [75 ,25])
    iqr = q75 - q25

    min_val = q25 - (iqr*1.5)
    max_val = q75 + (iqr*1.5)
    data_HCM = data_HCM[(data_HCM[variable] <= max_val) & (data_HCM[variable] >= min_val)]
```

```python
fig,((ax1,ax2,ax3), (ax4,ax5,ax6), (ax7,ax8,ax9), (ax10,ax11,ax12)) = plt.subplots(figsize=(20,15),ncols=3,nrows=4)
textprops = {"fontsize":20}

sns.boxplot(x = data_HCM['WS50M'], ax = ax1)
sns.boxplot(x = data_HCM['WS50M_RANGE'], ax = ax2)
sns.boxplot(x = data_HCM['QV2M'], ax = ax3)
sns.boxplot(x = data_HCM['PRECTOTCORR'], ax = ax4)
sns.boxplot(x = data_HCM['T2M_RANGE'], ax = ax5)
sns.boxplot(x = data_HCM['ALLSKY_SFC_SW_DWN'], ax = ax6)
sns.boxplot(x = data_HCM['T2M'], ax = ax7)
sns.boxplot(x = data_HCM['ALLSKY_KT'], ax = ax8)
sns.boxplot(x = data_HCM['ALLSKY_SFC_LW_DWN'], ax = ax9)
sns.boxplot(x = data_HCM['WS2M'], ax = ax10)
sns.boxplot(x = data_HCM['ALLSKY_SFC_PAR_TOT'], ax = ax11);
```
